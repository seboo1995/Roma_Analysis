## Doing TF_IDF for the corpus
### PREPORCESSING

```python
#lowercase
def to_lower_case(j):
    return j.text.lower()
df.text = df.apply(to_lower_case,axis=1)
```

```python
corpus = list(df.text)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
print(stopwords.words("english"))
```

```python
def remove_stopwords(j):
    temp = []
    spl = j.text.split(" ")
    filtered_words = [word for word in spl if word not in stopwords.words('english')]
    return ' '.join([str(elem) for elem in filtered_words]) 
df.text = df.apply(remove_stopwords,axis=1)
```

```python
corpus = list(df.text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
```

```python
df_tdif = pd.DataFrame(X[0].T.todense(), index=vectorizer.get_feature_names(),columns=["TF-IDF"])
```

```python
df_tdif = df_tdif.sort_values("TF-IDF",ascending=False)
```

```python
plt.figure(figsize=(10,10))
df_tdif["TF-IDF"].nlargest(20).plot(kind='barh')
```
